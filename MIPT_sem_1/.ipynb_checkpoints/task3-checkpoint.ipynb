### Задача 4. Несмещенность оценок.

Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$.

**Вопрос:** Какие из этих оценок являются несмещенными? Поясните свой ответ. При возникновении затруднений см. задачу 1.

**Ответ:** Как было показано в вычисления $S^2$ -- смещенная ровно на множитель $\frac{n}{n-1}$. Cледовательно оценка $\frac{n}{n-1}S^2$ -- несмещенная.

Для каждой из приведенных выше оценок $\widehat{\theta}$ выполните следующие действия.

* Вычислите $k = 500$ независимых оценок $\widehat{\theta_1}, ... , \widehat{\theta_k}$ по независимым выборкам $(X_1^1, ... , X_n^1), ... , (X_1^k, ... , X_n^k)$, сгенерированным из распределения $U[0, 1]$. Далее вычислите среднее этих оценок, которое обозначим $\overline{\theta}$.

* Визуализируйте полученные значения, построив на **одном** графике точки $(\widehat{\theta_1}, $y$), ... , (\widehat{\theta_k}, y)$ и среднее *этих оценок* $(\overline{\theta}, y)$, где $y$ &mdash; произвольные различные (например 0, 1) координаты для трёх различных типов оценок ($S^2, \frac{n}{n-1}S^2$).

* Повторите действие четыре раза для $n \in \{5, 10, 100, 500\}$. В итоге получится четыре графика для различных $n$, на каждом из которых изображено поведение четырёх типов оценок и их среднее.

Используйте данный шаблон для визуализации значений:

```python
# Для каждой оценки:
plt.scatter(<независимые оценки> , np.zeros(k) + y,
alpha=0.1, s=100, color=<цвет>, label=<метка>)
plt.scatter(<независимые оценки>.mean(), y, marker='*', s=200,
color='w', edgecolors='black')

# Для всего графика:
plt.vlines(1, <наименьший y>, <наибольший y>, color='r')
plt.title('sample size = %d' % k)
plt.yticks([])
plt.legend()
```

**Решение:**

```python
# Для каждой оценки:
for n in [5,10,100,500]:
    plt.figure(figsize=(8, 3))
    k = 500
    sample = sps.norm.rvs(size=(500,n))
    varss = sample.var(axis= 1)
    y = [0,1]
    plt.scatter(varss , np.zeros(k) + y[0], 
            alpha=0.1, s=100, color="cyan", label='$S^2$')
    plt.scatter(varss.mean(), y[0], marker='*', s=200, 
            color='w', edgecolors='black')
    plt.scatter(varss* n/(n-1) , np.zeros(k) + y[1], 
            alpha=0.1, s=100, color="yellow", label='$\dfrac{n}{n-1} S^2$')
    plt.scatter(varss.mean() * n/(n-1), y[1], marker='*', s=200, 
            color='w', edgecolors='black')

# Для всего графика:
    #plt.vlines(1, 0, 1, color='r')
    plt.title('sample size =' + str(k) +', n = %d' %n)
    plt.yticks(y)
    plt.legend()
    plt.show()
```

### Задача 5. Гамма-излучение.

Предлагается изучить некоторые свойства распределения Коши с параметром сдвига $\theta$, обладающего плотностью распределения $p_{\theta}(x) = \frac{1}{\pi \left(1 + \left(x- \theta\right)^2\right)}$.

*Замечание:* Такое распределение встречается, к примеру, в следующей задаче. На высоте 1 метр от точки $\theta$ находится источник $\gamma$-излучения, 
причем направления траекторий $\gamma$-квантов случайны, т.е. равномерно распределены по полуокружности. Тогда $X_i, i=1,...,n$ — зарегистрированные координаты точек пересечения $\gamma$-квантов с поверхностью детекторной плоскости — образуют выборку из распределения Коши со сдвигом $\theta$.

**1.** На отрезке $[-7, 7]$ постройте плотность стандартного нормального распределения и стандартного распределения Коши. Не забудьте добавить подписи к графику.

**Решение:**

```python
grid = np.linspace(-7, 7, 1000)

plt.figure(figsize=(10, 5))
plt.plot(grid, sps.cauchy.pdf(grid), color="green", label='Коши', lw=2)
plt.plot(grid, sps.norm.pdf(grid), color="orange", 
         label='$\mathcal{N}(0, 1)$', lw=2)
plt.xticks(fontsize=16), plt.yticks(fontsize=16)
plt.title('Плотности распределений', fontsize=16)
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.show()
```

**Вывод:** Коши имеет меньший максимум и более "тяжелые" концы.

**2.** Сгенерируйте выборку $X = \left(X_1, \dots, X_{1000} \right)$ из стандартного распределения Коши ($\theta = 0$). Для всех $n \leqslant 1000$ по первым $n$ элементам выборки $X_1, \dots, X_n$ вычислите значения следующих оценок:
- $\overline{X}$  —  выборочное среднее;
- $\widehat{\mu}$ —  выборочная медиана;
- Oдношаговая оценка, построенная по выборочной медиане (см. лекцию 4). 

В случае выборочной медианы можно использовать цикл по подвыборкам.

На одном графике изобразите зависимость значений этих оценок от $n$.
Сравните асимптотические дисперсии оценок (для тех оценок, для которых они существуют). Сделайте вывод.

**Замечание:** если некоторые оценки имеют большой разброс, и разница между графиками зависимостей оценок с малыми значениями недостаточно заметна, стоит сделать два графика, на одном из которых будут изображены все оценки, а на втором &mdash; только достаточно хорошие (мало отличающиеся от оцениваемого параметра), то есть этот график будует иметь меньший масштаб по оси $y$.

**Решение:**

```python
n=10000
sample_cauchy = sps.norm.rvs(size=(n))
means = sample_cauchy.cumsum() / np.arange(1,n+1)
median = []
for i in range(n):
    median.append(np.median(sample_cauchy[:i + 1]))
median = np.array(median)

first_app = []
for i in range(n):
    d1 = ((sample_cauchy - median[i])/ (1+(sample_cauchy - median[i])**2)).sum()
    d2 = ((1- (sample_cauchy - median[i])**2)/ (1+(sample_cauchy - median[i])**2)**2).sum()
    first_app.append(median[i] + d1 /d2 ) 

first_app = np.array(first_app)
```

```python
plt.figure(figsize=(12, 7))

plt.plot(np.arange(1,n+1), means, color="purple", label='$\overline{X}$')

plt.plot(np.arange(1,n+1), median, color="red", label='\widehat{\mu}')

plt.plot(np.arange(1,n+1), first_app, color="green", label='Oдношаговая оценка')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
k=1000
plt.plot(np.arange(1,k+1), means[:k], color="purple", label='$\overline{X}$')

plt.plot(np.arange(1,k+1), median[:k], color="red", label='\widehat{\mu}')

plt.plot(np.arange(1,k+1), first_app[:k], color="green", label='Oдношаговая оценка')

plt.legend()
plt.show()
```

**Вывод:** Медиана как ивыборочное среднее сходятся довольно долго, а вот одношаговая оценка моментально.
