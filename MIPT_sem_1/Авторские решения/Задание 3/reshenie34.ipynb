###Задача 4.
 В этой задаче предлагается изучить *свойство несмещённости*.

**1.** Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$.

**Вопрос:** Какие из этих оценок являются несмещенными? Поясните свой ответ. При возникновении затруднений см. задачу 1.

**Ответ:** $\frac{n}{n-1}S^2$.

$\mathsf{E}_\sigma S^2 = \mathsf{E}_\sigma \overline{X^2} - \mathsf{E}_\sigma \overline{X}^2$ 
= $\frac{1}{n}\mathsf{E}_\sigma \sum\limits_{i=1}^{n}X_i^2 - \frac{1}{n^2}\mathsf{E}_\sigma \sum\limits_{i=1}^{n} X_i^2 - \frac{1}{n^2}\sum\limits_{i \neq j} \mathsf{E}_\sigma X_i X_j$ 
= $\frac{n-1}{n}\mathsf{E}_\sigma X^2 - \frac{n(n-1)}{n^2} \mathsf{E}_\sigma X^2$
= $\frac{n-1}{n}\mathsf{D}_\sigma X$

$\Rightarrow$ $S^2$ &mdash; смещённая оценка (асимптотически &mdash; несмещённая), $\frac{n}{n-1}S^2$ &mdash; несмещённая оценка $\sigma^2$, её также называют исправленной выборочной дисперсией. 

Для каждой из приведенных выше оценок $\widehat{\theta}$ выполните следующие действия.

* Вычислите $k = 500$ независимых оценок $\widehat{\theta_1}, ... , \widehat{\theta_k}$ по независимым выборкам $(X_1^1, ... , X_n^1), ... , (X_1^k, ... , X_n^k)$, сгенерированным из распределения $U[0, 1]$. Далее вычислите среднее этих оценок, которое обозначим $\overline{\theta}$.

* Визуализируйте полученные значения, построив на **одном** графике точки $(\widehat{\theta_1}, $y$), ... , (\widehat{\theta_k}, y)$ и среднее *этих оценок* $(\overline{\theta}, y)$, где $y$ &mdash; произвольные различные (например 0, 1) координаты для трёх различных типов оценок ($S^2, \frac{n}{n-1}S^2$).

* Повторите действие четыре раза для $n \in \{5, 10, 100, 500\}$. В итоге получится четыре графика для различных $n$, на каждом из которых изображено поведение двух типов оценок и их среднее.

Используйте данный шаблон для визуализации значений:

```python
# Для каждой оценки:
plt.scatter(<независимые оценки> , np.zeros(k) + y, 
            alpha=0.1, s=100, color=<цвет>, label=<метка>)
plt.scatter(<независимые оценки>.mean(), y, marker='*', s=200, 
            color='w', edgecolors='black')

# Для всего графика:
plt.vlines(1, <наименьший y>, <наибольший y>, color='r')
plt.title('sample size = %d' % k)
plt.yticks([])
plt.legend()
```

**Решение:**

Напишем функции, вычисляющие оценки из условия и принимающие на вход выборку.

```python
def estimate_s2(X):
    return (X * X).mean() - (X.mean()) ** 2

def estimate_s2_corrected(X):
    n = X.size
    return estimate_s2(X) * (n) / (n - 1.)
```

Для каждой оценки сделаем кортеж вида:

(функция вычисляющая оценку, обозначение оценки в LaTex, цвет, y-координата для графика)

Заведём список таких кортежей для всех оценок.

```python
estimators = [
    (estimate_s2, "$S^2$", "red", 0.0),
    (estimate_s2_corrected, "$\\frac{n}{n-1}S^2$", "green", 0.1)
]
```

Напишем функцию, по распределению, списку оценок в формате, описанном выше, списку n и параметру k строит требуемый график.

```python
def est_plot(distribution, estimators, n_list, k):  
    """ estimators - список троек вида 
    (функция_вычисляющая_оценку_по_выборке, "её_название", 
    "eё цвет", y-координата)
    """
    
    for n in n_list:
        X = distribution.rvs(size=(k, n))
        plt.figure(figsize=(14, 0.5 * len(estimators)))
        for estimator, name, color, y in estimators:
            E = np.array([estimator(Xn) for Xn in X])

            plt.scatter(E, np.zeros(k) + y, alpha=0.1,
                        s=100, color=color, label=name)
            plt.scatter(E.mean(), y, marker='*', s=200,
                        color='w', edgecolors='black')

        plt.vlines(1, estimators[0][-1], estimators[-1][-1], color='r')
        plt.title('sample size = %d' % n)
        plt.yticks([])
        plt.legend()
        plt.xlim((0.6, 1.4))
        plt.show()
```

Инициализируем остальные параметры и построим график.

```python
n_list = (10, 100, 500)
k = 500

norm_distibution = sps.norm(loc=0., scale=1.) 
est_plot(norm_distibution, estimators=estimators, 
         n_list=n_list, k=k)
```

**2.** Изучим поведение среднего оценок из первого пункта при росте размера $n$ выборки. Постройте график зависимости $\overline{\theta}$ от $n$ для трёх типов оценок. Для вычисления зависимости нужно один раз сгенерировать выборки из п. 1.1 достаточно большого размера и посчитать оценки по префиксам, используя функции из `numpy` и не используя циклы. Какие из оценок являются асимптотически несмещёнными (т.е. $\forall \theta \in \Theta\colon \mathsf{E}_\theta \widehat{\theta} \to \theta$ при $n\to +\infty$)?

**Решение:**

Обе оценки являются асимптотически несмещенными.

Напишем функцию, строющую требуемый график, получающую на вход те же параметры, что и в первой части.

```python
def mean_plot(distribution, estimators, n_grid, k):
    mean_values = [
        [
            np.mean([
                estimate(distribution.rvs(size=int(n))) for _ in range(k)
            ]) for n in n_grid
        ] for estimate, _, _, _ in estimators
    ]
    
    plt.figure(figsize=(10, 7))
    plt.grid()
    for mean_values, (_, name, color, _) in zip(mean_values, estimators):
        plt.plot(n_grid, mean_values, label = name, color=color)
    plt.xlabel("n")
    plt.ylabel("Среднее оценок")
    plt.title("Зависимость среднего оценок от n")
    plt.legend()
    plt.show()
```

```python
%%time
n_grid = np.linspace(10, 300, 100)
mean_plot(norm_distibution, estimators, n_grid, k)
```

**Вывод:** Как отмечено выше, все оценки являются асимптотически несмещёнными.

Сделайте вывод о том, что такое свойство несмещенности. Подтверждают ли сделанные эксперименты свойство несмещенности данных оценок? Поясните, почему в лабораторных по физике при оценке погрешности иногда используют $n-1$ в знаменателе, а не $n$.

**Вывод:** несмещённость &mdash; свойство оценки в **среднем** (матожидание) быть равной параметру, который она оценивает. 


 Оценка $s^2$ является только асимптотически несмещенной, а $\frac{n}{n-1}s^2$ является как асимптотически несмещенной, так и просто несмещенной. Проведённые эксперименты подтверждают эти свойства.
