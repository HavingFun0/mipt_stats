```python
import torch
import matplotlib.pyplot as plt

from torchviz import make_dot
```

### Задача 1

Изобразите график параметрически заданной функции. Сами функции и переменные необходимо задавать с помощью Pytorch.

$$ x=4.4\times\left(\cos(t)+\frac{\cos(1.1\times{t})}{1.1}\right)$$

$$y=4.4\times\left(\sin(t)−\frac{\sin(1.1\times{t})}{1,1}\right)$$

$$t∈[0;20π]$$ 

```python
plt.figure(figsize=(10, 7))

lin_sp = torch.linspace(0, 20*torch.pi, 100000)

x = 4.4 * (torch.cos(lin_sp)+torch.cos(lin_sp*1.1)/1.1)
y = 4.4 * (torch.sin(lin_sp)-torch.sin(lin_sp*1.1)/1.1)

plt.plot(x.numpy(), y.numpy())

plt.title('Параметрический график')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### Задача 2

Для функции

$$ f(x, y, z)=\left(\frac{y^4}{1+e^{-x}}\right)^3 + z$$

С помощью Pytorch: 

1. Постройте вычислительный граф с возможностью считать производные по x, y.

2. Выведите все узлы полученного графа и их атрибуты (рассмотренные на лекции)

3. Объясните, почему атрибуты принимают такие значения 

```python
# Пункт 1
x = torch.tensor(1.0, requires_grad=True)
y = torch.tensor(1.0, requires_grad=True)
z = torch.tensor(1.0, requires_grad=False)

f = (y**4/(1+torch.exp(-x)))**3 + z
f
```

```python
params = dict([('x', x), ('y', y),('z',z)])
params
```

```python
make_dot(f, params=params, show_attrs=True)
```

```python
# Пункт 2

# Displaying
for i, name in zip([x, y, z, f], "xyzf"):
    print(f"{name}\ndata: {i.data}\nrequires_grad: {i.requires_grad}\n\
grad: {i.grad}\ngrad_fn: {i.grad_fn}\nis_leaf: {i.is_leaf}\n")
```

Переменные x,y,z являются листами по определению, а f(x,y,z) не является, т.к. она зависит от переменной z, по которой нельзя дифференцировать.

### Задача 3

В условиях задачи 2 в точке (1, 3, 2):

1. Посчитайте для функции все возможные первые частные производные методом backward()
2. Посчитайте для функции все возможные вторые частные производные 

**Пункт 1**


```python
x = torch.tensor(1.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)
z = torch.tensor(2.0, requires_grad=False)

f = (y**4/(1+torch.exp(-x)))**3 + z
```

```python
f.backward()

print(f'df/dx = {x.grad.item()}\ndf/dy = {y.grad.item()}')
```

**Пункт 2**

Возможные вторые производные: $dx^2 ,~ dxdy, ~dydx, ~dy^2$.

```python
from torch.autograd import grad
```

```python
f = (y**4/(1+torch.exp(-x)))**3 + z

f_x = grad(outputs = f, inputs = x, retain_graph  = True, create_graph  = True)
f_y = grad(outputs = f, inputs = y, retain_graph  = True, create_graph  = True)
```

```python
f_x_x = grad(outputs = f_x, inputs = x, retain_graph  = True, create_graph  = True)
print('d^2f/dxdx = '+str(f_x_x[0].item()))


f_y_y = grad(outputs = f_y, inputs = y, retain_graph  = True, create_graph  = True)
print('d^2f/dxdy = '+str(f_y_y[0].item()))


f_y_x = grad(outputs = f_x, inputs = y, retain_graph  = True, create_graph  = True)
print('d^2f/dydx = '+str(f_y_x[0].item()))


f_x_y = grad(outputs = f_y, inputs = x, retain_graph  = True, create_graph  = True)
print('d^2f/dydy = '+str(f_x_y[0].item()))
```
